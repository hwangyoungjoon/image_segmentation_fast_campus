# mnist_classification_using_ANN

```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

- 데이터 준비 및 one hot encoding

```python
mnist=input_data.read_data_sets('/tmp/data',one_hot=True)
```

```python
# 학습을 위한 파라미터 튜닝
learning_rate=0.001
training_epochs=30 # 학습횟수
batch_size=256     # 미니배치할 사이즈
display_step=1     # 손실함수 출력 횟수
```

```python
# 네트워크 구조 지정을 위한 파라미터들
input_size=784  # 그림이 (28*28*1)이다.
hidden1_size=256   # 히든 레이어1의 노드 갯수
hidden2_size=256   # 히든 레이어2의 노드 갯수
output_size=10     # one hot 인코딩 떄문에 사이즈가 10 이다
```

```python
# 입력값과 출력값을 받기위한 플레이스 홀더
x=tf.placeholder(tf.float32,[None,input_size])  # None은 28*28*1 이미지를 몇장 받을지 상관없다.는 것을 의미
true_y=tf.placeholder(tf.float32,[None,output_size])
```

```python
# ANN 모델 구현
def build_ANN(x):
    w1=tf.Variable(tf.random_normal([input_size,hidden1_size]))
    b1=tf.Variable(tf.random_normal([hidden1_size]))
    L1_output=tf.nn.relu(tf.matmul(x,w1)+b1)
    
    w2=tf.Variable(tf.random_normal([hidden1_size,hidden2_size]))
    b2=tf.Variable(tf.random_normal([hidden2_size]))
    L2_output=tf.nn.relu(tf.matmul(L1_output,w2)+b2)
    
    w_output=tf.Variable(tf.random_normal([hidden2_size,output_size]))
    b_output=tf.Variable(tf.random_normal([output_size]))
    output=tf.matmul(L2_output,w_output)+b_output
    
    return output


# 모델 구현
predicted_value= build_ANN(x)
```

```python
# 손실함수와 최적화 정의
loss=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=predicted_value,labels=true_y))
optimizer=tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(loss)
```

```python
# 세션 열고 그래프 실행
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer()) #변수들에 초기값을 할당한다.
    
    for epoch in range(training_epochs):
        average_loss=.0
        total_batch=int(mnist.train.num_examples/batch_size)  # 전체 배치를 불러온다
        for i in range(total_batch):  # 한 번 트레이닝에 256개씩 트레이닝하고 한 에폭을 도는데 214번이 걸린다
            batch_x,batch_y=mnist.train.next_batch(batch_size)  
            #옵티마이저를 실행해서 파라미터들을 업데이트 한다.
            _, current_loss=sess.run([optimizer,loss],feed_dict={x:batch_x,true_y:batch_y})
            #평균 손실을 측정한다.
            average_loss+=current_loss/total_batch
            # 지정된 epoch마다 학습결과를 보여준다. 
            if epoch % display_step==0:
                print("epoch",(epoch+1),"손실 함수(loss)=","{:.9f}".format(average_loss))
    print("학습 끝")
    
    
    #테스트를 통해 얼마나 정확한지 출력한다
    correct_predicition=tf.equal(tf.argmax(predicted_value,1),tf.argmax(true_y,1))
    accuracy=tf.reduce_mean(tf.cast(correct_predicition,'float'))
    print("정확도:",accuracy.eval({x:mnist.test.images,true_y:mnist.test.labels}))
```
